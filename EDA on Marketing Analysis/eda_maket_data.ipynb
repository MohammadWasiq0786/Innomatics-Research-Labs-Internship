````python
df.columns= df.columns.str.replace(' ', '')
df.columns
````

### Data Cleaning
* **Remove the Special Characters**
* **Remove the Spaces**
* **Indentifier Name**
* **Change the data type**

````python
# Convert Income feature to numerical
print("Income Column before cleaning", '\n', df['Income'].head())
````

````python
df['Income']= df['Income'].str.replace('$', '')
df['Income']= df['Income'].str.replace(',', '')
df['Income']= df['Income'].astype('float')

print("Income Column after cleaning", '\n', df['Income'].head())
````

````python
df.info()
````

Now the **Income** column is clean and converted into `float` data type.



````python
df.isnull().sum()
````

````python
df.isnull().sum().sort_values(ascending= False)
````

````python
# Fill the missing values
df['Income']= df['Income'].fillna(df['Income'].median())
df.isnull().sum().sort_values(ascending= False)
````

**Now there is no missing value in our dataset.**

````python
# Treatment of outliers
clean_df= df[df.Income <= 600000]

clean_df['Income'].plot(kind= 'box')
````

````python
# 1.5 * IQR
q1= df['Income'].quantile(0.25)
q3= df['Income'].quantile(0.75)

IQR= q3-q1

income_lower_bound= q1 - 1.5*IQR
income_upper_bound= q3 + 1.5*IQR

clean_df_2= df[(df['Income'] >= income_lower_bound) & (df['Income'] <= income_upper_bound)]

.clean_df_2['Income'].plot(kind= 'box')
````

**Now there is no outlier in our data.**
